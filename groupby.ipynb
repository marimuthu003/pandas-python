```python
#groupby method:

#splitting first,
#applying function,
#combine those results.

country_grp=df.groupby(['Country'])
```

```python
#by using filter method:
filt=df['Country']=='Brazil'
df.loc[filt,'DevType'].value_counts()
```

```python
#by using country group instead of filter:

country_grp['DevType'].value_counts().loc['Brazil']
```

```python
country_grp['DevType'].value_counts(normalize=True).loc['Brazil']
```

```python
country_grp['ConvertedCompYearly'].median().loc['Brazil']
```

```python
country_grp['ConvertedCompYearly'].agg(['median','mean']).loc['Brazil']
```

```python
#for summing results in filtering:
filt=df['Country']=='Brazil'
df.loc[filt,'LanguageHaveWorkedWith'].str.contains('python').value_counts().sum()
```

```python
#for summing in country group use apply( method instead of filltering):

country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('HTML/CSS').sum())
```

```python
#for caculating percentage :
country_respondents=df['Country'].value_counts()
country_respondents=pd.DataFrame(country_respondents)
country_respondents
```

```python
country_uses_htmlcss=country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('HTML/CSS').sum())
country_uses_htmlcss=pd.DataFrame(country_uses_htmlcss)
country_uses_htmlcss
```

```python
htmlcss_df=pd.concat([country_respondents,country_uses_htmlcss],axis='columns',sort=False)
htmlcss_df
```

```python
htmlcss_df.rename(columns={'count':'countrycount','LanguageHaveWorkedWith':'languagecount'})
```

```python
#calculating percentage:

htmlcss_df['percentage']=(htmlcss_df['countrycount']/htmlcss_df['languagecount']*100)
htmlcss_df
```
